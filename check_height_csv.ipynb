```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa

def plot_heightmap(csv_path, delimiter=",", stride=10, cmap="viridis"):
    # --- CSV 로드 ---
    Z = np.loadtxt(csv_path, delimiter=delimiter, dtype=np.float32)
    h, w = Z.shape
    X, Y = np.meshgrid(np.arange(w), np.arange(h))

    print(f"Shape: {Z.shape}, dtype={Z.dtype}")
    print(f"3D stride: {stride}")

    # --- 2D 히트맵 ---
    plt.figure(figsize=(7,6))
    im = plt.imshow(Z, cmap=cmap, origin="upper")
    plt.colorbar(im, label="Height")
    plt.title("Heightmap - 2D view")
    plt.xlabel("X (pixel)")
    plt.ylabel("Y (pixel)")
    plt.show()

    # --- 3D 서피스 ---
    fig = plt.figure(figsize=(9,7))
    ax = fig.add_subplot(111, projection="3d")
    ax.plot_surface(X[::stride, ::stride],
                    Y[::stride, ::stride],
                    Z[::stride, ::stride],
                    cmap=cmap, linewidth=0, antialiased=False)
    ax.set_xlabel("X (pixel)")
    ax.set_ylabel("Y (pixel)")
    ax.set_zlabel("Height")
    plt.title("Heightmap - 3D Surface")
    plt.show()

```

```python
# 사용 예시
plot_heightmap("250814_170803_height.csv", delimiter=",", stride=20)
```
